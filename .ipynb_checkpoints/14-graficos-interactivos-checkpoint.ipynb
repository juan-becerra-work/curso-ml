```python
import pandas as pd
import numpy as np
import cufflinks as cf
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
```

```python
init_notebook_mode(connected=True)
cf.go_offline()
# directiva para poder visualizar los gráficos dentro del notebook
%matplotlib inline
```

```python
dataframe = pd.DataFrame(np.random.randn(100, 4), columns=['a', 'b', 'c', 'd'])
dataframe
```

```python
dataframe.plot()
```

```python
# DA ERROR POR FALTA DE ATRIBUTO HoverinfoValidator EN MÓDULO plotly.validators.carpet

dataframe.iplot()
```

```python
# DA ERROR POR FALTA DE ATRIBUTO HoverinfoValidator EN MÓDULO plotly.validators.carpet

dataframe.iplot(kind='scatter', x='a', y='b', mode='markers')
```
