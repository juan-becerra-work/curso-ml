```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Dataset incluido en Seaborn
propinas =sns.load_dataset('tips')
propinas.head(10)
```

```python
# Distribución
sns.distplot(propinas['total_bill'])
```

```python
# Más detallado
sns.distplot(propinas['total_bill'], bins=30) # Bins indica, en este caso, la cantidad de barras
```

```python
# Comparar dos variables
sns.jointplot(x='total_bill', y='tip', data=propinas) # Por default, scatter
```

```python
# Comparar dos variables
sns.jointplot(x='total_bill', y='tip', data=propinas, kind='hex') # Hexágonos
```

```python
# Comparar dos variables
sns.jointplot(x='total_bill', y='tip', data=propinas, kind='reg') # Regresión lineal
```

```python
# Comparar dos variables
sns.jointplot(x='total_bill', y='tip', data=propinas, kind='kde') # KDE
```

```python
# Comparar todas las variables numéricas de a pares
sns.pairplot(propinas) # pair plot
```

```python
# Comparar con categorías
sns.pairplot(propinas, hue='sex') # Métricas por sexo del que paga
```

```python
# Cada dato en el dataset
sns.rugplot(propinas['total_bill'])
```

## Gráficos de categorías

```python
# Barras
sns.barplot(x='sex', y='total_bill', data=propinas)
```

```python
# Contar miembros de categorías
sns.countplot(x='sex', data=propinas)
```

```python
# Comparar una categoría con una columna no numérica
sns.boxplot(x='day', y='total_bill', data=propinas)
```

```python
# Comparar dos categorías con una columna no numérica
sns.boxplot(x='day', y='total_bill', data=propinas, hue='smoker')
```

```python
# Comparar dos categorías con una columna no numérica (variante)
sns.violinplot(x='day', y='total_bill', data=propinas, hue='smoker')
```

```python
# Comparar dos categorías con una columna no numérica (variante, juntando ambas categorías)
sns.violinplot(x='day', y='total_bill', data=propinas, hue='smoker', split=True)
```

```python
# Mostrar todos los puntos del dataset
sns.stripplot(x='day', y='total_bill', data=propinas, hue='smoker', dodge=True) # dodge los junta o separa por categoría hue
```

```python
# Mostrar todos los puntos del dataset, evitando que se solapen
sns.swarmplot(x='day', y='total_bill', data=propinas, hue='smoker', dodge=True) # dodge los junta o separa por categoría hue
```

```python
# Mezclar gráficos
sns.violinplot(x='day', y='total_bill', data=propinas)
sns.swarmplot(x='day', y='total_bill', data=propinas)
```

## Mapas de calor

```python
# Dataset incluido en Seaborn
vuelos =sns.load_dataset('flights')
vuelos.head(10)
```

```python
# Para un mapa de calor (heat map), se necesita tener la tabla expresada como matriz
vuelos_matrix = vuelos.pivot_table(index='month', columns='year', values='passengers')
vuelos_matrix
```

```python
# Gráfico mapa de calor
sns.heatmap(vuelos_matrix)
```

```python
# Colores
sns.heatmap(vuelos_matrix, cmap='coolwarm')
```

```python
# Generar espacios
sns.heatmap(vuelos_matrix, cmap='coolwarm', linecolor='white', linewidths=1)
```

## Cuadrículas

```python
# Dataset incluido en Seaborn
flores =sns.load_dataset('iris')
flores.head(10)
```

```python
sns.pairplot(flores)
```

```python
# Elegir el tipo de gráfico de comparación
# Pair Grid
graficos = sns.PairGrid(flores)
graficos.map(plt.scatter)
```

```python
# Elegir el tipo de gráfico de comparación
graficos = sns.PairGrid(flores)
graficos.map_diag(sns.distplot) # Gráficos de la diagonal
graficos.map_upper(plt.scatter) # Gráficos encima de la diagonal
graficos.map_lower(sns.kdeplot) # Gráficos debajo de la diagonal
```

```python
# Facet Grid

# Dataset incluido en Seaborn
propinas =sns.load_dataset('tips')
propinas.head(10)
```

```python
propinas['time'].unique()
```

```python
graficos2 = sns.FacetGrid(data=propinas, col='time', row='smoker')
graficos2.map(sns.distplot, 'total_bill')
```

```python
# Cambiar el tipo de gráfico
graficos2 = sns.FacetGrid(data=propinas, col='time', row='smoker')
graficos2.map(plt.scatter, 'total_bill', 'tip')
```

## Regresión

```python
propinas.head(10)
```

```python
sns.lmplot(x='total_bill', y='tip', data=propinas)
```

```python
# Agregar una categoría
sns.lmplot(x='total_bill', y='tip', data=propinas, hue='sex', markers=['o','X'])
```

```python
# Ampliar tamaño
sns.lmplot(x='total_bill', y='tip', data=propinas, hue='sex', markers=['o','X'], scatter_kws={'s': 80})
```

```python
# Separar por filas y columnas
sns.lmplot(x='total_bill', y='tip', data=propinas, col='day', row='sex')
```

```python
# Separar dentro del mismo gráfico por filas y columnas
sns.lmplot(x='total_bill', y='tip', data=propinas, col='day', hue='sex')
```

```python
# Modificar tamaño
sns.lmplot(x='total_bill', y='tip', data=propinas, aspect=0.8, height=10)
```
