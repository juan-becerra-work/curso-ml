```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(2, 9, 10)
y = x**2
```

```python
plt.plot(x, y, 'red') # En Jupyter no se requiere el método show
plt.title('Gráfico')
plt.xlabel('Eje x')
plt.ylabel('Eje y')
plt.show()
```

## Más de un gráfico en la misma área

```python
# En 2 columnas
plt.subplot(1, 2, 1)
plt.plot(x, y, 'g')

plt.subplot(1, 2, 2)
plt.plot(y, x, 'blue')

plt.show
```

```python
# En 2 filas
plt.subplot(2, 1, 1)
plt.plot(x, y, 'g')

plt.subplot(2, 1, 2)
plt.plot(y, x, 'blue')

plt.show
```

```python
# Usando OOP
figura = plt.figure()
grafico = figura.add_axes([0.1, 0.1, 0.8, 0.8]) # Tamaño en %
grafico.plot(x, y, 'green')
```

```python
# Usando OOP, 2 gráficos
figura = plt.figure()
grafico1 = figura.add_axes([0.1, 0.1, 0.8, 0.8])
grafico1.plot(x, y, 'green')

grafico2 = figura.add_axes([0.2, 0.5, 0.3, 0.3])
grafico2.plot(y, x, 'red')
```

## Multigráficos

```python
figura, graficos = plt.subplots(nrows=2, ncols=2)

graficos[0][0].plot(x, y, 'blue')
graficos[0][1].plot(x, y, 'green')
graficos[1][0].plot(x, y, 'yellow')
graficos[1][1].plot(x, y, 'red')

graficos[0][0].set_title('Gráfico 1')
graficos[0][1].set_title('Gráfico 2')
graficos[1][0].set_title('Gráfico 3')
graficos[1][1].set_title('Gráfico 4')
```

## Tamaños de figuras y gráficos

```python
figura = plt.figure(figsize=(8, 4) )
grafico = figura.add_axes([0.1, 0.1, 0.9, 0.9])
grafico.plot(x, y, 'g')
```

# Crear 2 gráficos en la misma figura

```python
figura = plt.figure()
graficos = figura.add_axes([0.1, 0.1, 0.9, 0.9])
graficos.plot(x, x**2, 'red', label='X al cuadrado')
graficos.plot(x, x**3, 'yellow', label='X al cubo')
graficos.legend(loc=0)
```

## Colores

```python
# https://matplotlib.org/tutorials/colors/colors.html
# https://htmlcolorcodes.com/

figura = plt.figure()
grafico = figura.add_axes([0, 0, 1, 1])
#grafico.plot(x, y, 'teal')
grafico.plot(x, y, '#1BF0D9')
```

## Tipo de línea

```python
# https://matplotlib.org/gallery/lines_bars_and_markers/line_styles_reference.html
figura = plt.figure()
grafico = figura.add_axes([0, 0, 1, 1])
grafico.plot(x, y, color='red', linewidth=5, alpha=0.5, linestyle=':')
```

# Marcadores

```python
figura = plt.figure()
grafico = figura.add_axes([0, 0, 1, 1])
grafico.plot(x, y, color='red', marker='X', markersize=10)
```
